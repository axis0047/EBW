```python
dataset = []
column_names = ['player', 'hand', 'dominant_hand', 'direction', 'overall_attempt_no', 'attempt_by_the_hand', 'intention',
                'performance', 'fail_time', 'time_to_complete_in_ms', 'no_of_touches', 'no_of_clusters', 'levene_test_p_val', 'levene_test_statistics', 'largest_cluster_size',
                'smallest_cluster_size', 'average_cluster_size', 'variance_of_cluster_size', 'max_inter_cluster_time', 'min_inter_cluster_time',
                'average_inter_cluster_time', 'variance_of_inter_cluster_time', 'max_inter_touch_time', 'min_inter_touch_time', 'average_inter_touch_time', 'variance_of_inter_touch_time']

#Method of getting set of values for k - using a left screwed distribution to include a lot of small values and a less of larger values

def skewed_values(num_values, range_start, range_end, skewness=2):
    values = np.random.exponential(scale=skewness, size=num_values)
    values = values / np.max(values)  # Normalize to the range [0, 1]
    values = values * (range_end - range_start) + range_start
    return values

#Method of getting optimal k

def optimal_k(inertia_array, k_val_array):

  ratio_array = []

  for i in range (1,len(inertia_array)):
    if inertia_array[i] != 0:
      ratio = inertia_array[i-1]/inertia_array[i]
      ratio_array.append(ratio)

  for i in range(0, len(ratio_array)):
    if ratio_array[i] < 1.9:
      return k_val_array[i]

#calculating inter touch time

def inter_touch(y_array):
  itt_array = []
  for i in range(1,len(y_array)):
    itt_array.append(y_array[i] - y_array[i -1])
  print('itt array',itt_array)
  itt_return_array = []
  itt_return_array.append(max(np.concatenate(itt_array)))
  itt_return_array.append(min(np.concatenate(itt_array)))
  itt_return_array.append(sum(np.concatenate(itt_array))/len(np.concatenate(itt_array)))
  if(len(itt_array) > 1):
    itt_return_array.append(statistics.variance(np.concatenate(itt_array)))
  else:
    itt_return_array.append(0)
  print('itt return array', itt_return_array)
  return itt_return_array

#calculating inter cluster time

def inter_cluster(cluster_array):
  print('cluster array000',cluster_array)

  sorted_cluster_array = []

  for cluster in cluster_array:
    sorted_cluster = np.sort(cluster)
    sorted_cluster_array.append(np.array(sorted_cluster))

  print('ssssss', sorted_cluster_array)

  print('arrrrrrrr',np.array((sorted_cluster_array), dtype=object))
  sorted_np_cluster_array = np.array((sorted_cluster_array) , dtype=object)
  print('snpca', sorted_np_cluster_array)
  sorted_sorted_cluster_array = sorted(sorted_np_cluster_array, key = lambda x:x[0])
  print('ssca',sorted_sorted_cluster_array)

  ict_array = []
  for i in range(1,len(sorted_sorted_cluster_array)):
    ict_array.append(sorted_sorted_cluster_array[i][0] - sorted_sorted_cluster_array[i-1][-1])
  # print(np.concatenate(ict_array))
  print('ict array',ict_array)

  ict_return_array = []
  ict_return_array.append(max(ict_array))
  ict_return_array.append(min(ict_array))
  ict_return_array.append(sum(ict_array)/len(ict_array))
  if(len(ict_array) > 1):
    ict_return_array.append(statistics.variance(ict_array))
  else:
    ict_return_array.append(0)

  return ict_return_array

#calculating cluster data

def cluster_size_data(cluster_array):
  cluster_size_array = []
  for cluster in cluster_array:
    cluster_size_array.append(len(cluster))
  print('cluster size array',cluster_size_array)
  cluster_size_return_array = []
  cluster_size_return_array.append(max(cluster_size_array))
  cluster_size_return_array.append(min(cluster_size_array))
  cluster_size_return_array.append(sum(cluster_size_array)/len(cluster_size_array))
  if(len(cluster_size_array) > 1):
    cluster_size_return_array.append(statistics.variance(cluster_size_array))
  else:
    cluster_size_return_array.append(0)
  return cluster_size_return_array


#doing thing

folder_path = "/content/gdrive/MyDrive/data_2"

file_list = os.listdir(folder_path)

inertia_array = []
k_val_array = []

try:

  for file_name in file_list:

      print(file_name)

      datarow = []
      file_name_splitted = file_name.split("_")

      file_path = os.path.join(folder_path, file_name)
      if os.path.isfile(file_path):

          with open(file_path, 'rb') as file:
              content = file.read()
              df = pd.read_csv(io.BytesIO(content))

              df = df.drop(df.index[0])
              df = df.drop(df.index[1])

              y = df['lapse-time']

              y_num = y.to_numpy()
              y_in = y_num.reshape(-1, 1)

              weight = 0

              if len(y) < 10:
                weight = 1
              elif len(y) < 20:
                weight = 0.8
              elif len(y) < 50:
                weight = 0.6
              else:
                weight = 0.4

              k_val_array = sorted(set(np.floor(np.array(sorted(set(skewed_values(math.floor(weight*len(y)), 0, len(y), 5))))).astype(int)))

              print(k_val_array)

              for k in k_val_array:

                if k != 0:
                  print(k)
                  kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
                  kmeans.fit(y_in)

                  cluster_centers = kmeans.cluster_centers_
                  labels = kmeans.labels_

                  inertia_array.append(kmeans.inertia_)

                  ratio_array = []

                  for i in range (1,len(inertia_array)):
                    if inertia_array[i] != 0:
                      ratio = inertia_array[i-1]/inertia_array[i]
                      ratio_array.append(ratio)

          op_k = optimal_k(inertia_array, k_val_array)

          print(op_k)

          kmeans_op = KMeans(n_clusters=op_k, init='k-means++', random_state=42)
          kmeans_op.fit(y_in)

          cluster_centers = kmeans_op.cluster_centers_
          labels_op = kmeans_op.labels_

          plt.scatter(y_in, np.zeros_like(y_in), c=labels_op, cmap='viridis', alpha=0.7, edgecolors='k')
          plt.scatter(cluster_centers, [0] * op_k, c='red', marker='X', s=100, label='Cluster Centers')

          plt.legend()
          plt.show()

          clusters = np.array([y_in[labels_op == i] for i in range(kmeans_op.n_clusters)], dtype = 'object')

          print("clusters in main", clusters)

          levene_array = []
          cluster_array = []

          for i in clusters:
            cluster_array = []
            for j in i:
              cluster_array.append(j[0])
            levene_array.append(cluster_array)

          print("levene array ", levene_array)

          statistic_levene, p_value_levene = stats.levene(*levene_array)

          itt_details = inter_touch(y_in)
          ict_details = inter_cluster(levene_array)
          cluster_data = cluster_size_data(levene_array)
          print('cluster data', cluster_data)

          if(len(file_name_splitted) == 9):

            datarow.append(file_name_splitted[1])
            datarow.append(file_name_splitted[2])
            datarow.append(file_name_splitted[0])
            datarow.append(file_name_splitted[6])
            datarow.append(file_name_splitted[3])
            datarow.append(file_name_splitted[4])
            datarow.append(file_name_splitted[5])
            datarow.append(file_name_splitted[5])
            datarow.append(file_name_splitted[8].split('.')[0])
            datarow.append(file_name_splitted[7])
            datarow.append(len(y_in))
            datarow.append(op_k)
            datarow.append(p_value_levene)
            datarow.append(statistic_levene)
            datarow.append(cluster_data[0])
            datarow.append(cluster_data[1])
            datarow.append(cluster_data[2])
            datarow.append(cluster_data[3])
            datarow.append(ict_details[0])
            datarow.append(ict_details[1])
            datarow.append(ict_details[2])
            datarow.append(ict_details[3])
            datarow.append(itt_details[0])
            datarow.append(itt_details[1])
            datarow.append(itt_details[2])
            datarow.append(itt_details[3])

          elif(len(file_name_splitted) == 8):

            datarow.append(file_name_splitted[1])
            datarow.append(file_name_splitted[2])
            datarow.append(file_name_splitted[0])
            datarow.append(file_name_splitted[6])
            datarow.append(file_name_splitted[3])
            datarow.append(file_name_splitted[4])
            datarow.append(file_name_splitted[5])
            datarow.append(file_name_splitted[5])
            datarow.append(0)
            datarow.append(file_name_splitted[7].split('.')[0])
            datarow.append(len(y_in))
            datarow.append(op_k)
            datarow.append(p_value_levene)
            datarow.append(statistic_levene)
            datarow.append(cluster_data[0])
            datarow.append(cluster_data[1])
            datarow.append(cluster_data[2])
            datarow.append(cluster_data[3])
            datarow.append(ict_details[0])
            datarow.append(ict_details[1])
            datarow.append(ict_details[2])
            datarow.append(ict_details[3])
            datarow.append(itt_details[0])
            datarow.append(itt_details[1])
            datarow.append(itt_details[2])
            datarow.append(itt_details[3])


          dataset.append(datarow)

      else:
        raise ValueError('File Not Found')

except Exception as e:
  traceback.print_exc()

output_dataframe = pd.DataFrame(dataset, columns = column_names)

output_dataframe.to_csv('EBW_intention_dataset.csv', index = False)
files.download('EBW_intention_dataset.csv')
```
